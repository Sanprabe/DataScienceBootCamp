```python
df = pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv', encoding='latin1')
df
```

```python
## Data Wrangling

df = df.dropna() # Varias filas vacias en el df
df
```

```python
## Primer Punto
#  Construir un objeto de series de tiempo con el índice igual al año de los juegos olímpicos

index_df = pd.to_datetime(df['Year'], format='%Y') # Se crea una Serie con los valores de la columna años
index_df = list(index_df) # Se convierte en una lista
df.index = index_df # Se usa la lista como index
del df['Year'] # Elimina la columna Year

df
```

```python
## Segundo Punto
#  Separar nombres de apellidos en dos columnas distintas usando la coma como separador

names = df.Athlete.str.split(",",expand=True) # Crear un Data Frame con valores que resultan de hacer split de cada atleta
firstName = names[1] # Crear serie con el nombre
lastName = names[0] # Crear serie con el apellido

df = df.assign(FirstName=firstName) # Crear nueva columna en el df con nombre FirstName
df = df.assign(LastName=lastName)  # Crear nueva columna en el df con nombre LastName

df
```

```python
## Tercer Punto
#  Obtener medidas resumen del conjunto de datos. Cuál es el país que ganó más medallas?

paisConMasMedallas = list(df['Country'].value_counts().head(1).index)[0]
print(f'País con más medallas: {paisConMasMedallas}')

df.describe()
```

```python
## Cuarto Punto
#  Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento.

df_mans_raw = df[df['Gender'] == 'Men']['Gender'] # Se hace un filtro para solo mostrar atletas hombres
male_winners_year = df_mans_raw.groupby(df_mans_raw.index).count() # Se agrupan por año y se cuentan el numero de medallas
male_winners_year
```
