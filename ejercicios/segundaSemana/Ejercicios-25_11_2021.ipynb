## Ejercicio 1

Input: ```list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]```

Obtener el siguiente output usando "append":

  Output: ```[10, 20, [300, 400, [5000, 6000, 7000], 500], 30, 40]```

## Ejercicio 2

Input1: ```s1 = {10, 20, 30, 40, 50}```

Input2: ```s2 = {30, 40, 50, 60, 70}```

Responder
1. Obtener el siguiente output con alguno de met√≥dos vistos

  Output1: ```{40, 50, 30}```

2. Como obtener el siguiente output con alguno de los met√≥dos vistos para sets

  Output2: ```{70, 40, 10, 50, 20, 60, 30}```

3. Extraer todos los elementos de s2 que no estan en s1

````python
s1 = {10, 20, 30, 40, 50}
s2 = {30, 40, 50, 60, 70}

print('Punto 1')
print(s2.intersection(s1))
print('-----------------')
print('Punto 2')
print(s1.union(s2))
print('-----------------')
print('Punto 3')
s2.symmetric_difference(s1)
````

## Ejercicio 3

input: ```tupla=(10, 20, 30, 40, 50)```

Responder:

1. Ordenar la tupla de mayor a menor
2. Extraer los valores de la tupla en las siguientes variables: a,b,c,d,e
3. Ahora suponga que tiene la siguiente tupla: 
```tupla_d = (11, [22, 33], 44, 55)```
Recordando el concepto de inmutabilidad de las tuplas, ser√≠a posible poder modificar el valor **22** por **222** o en este caso no se puede?. Justifiquen

````python
tupla=(10, 20, 30, 40, 50)
print('Punto 1')
print(tupla[::-1]) # No la organiza, pero genera una sorted list con los objetos en la tupla
print('-----------------')

print('Punto 2')
a, b, c, d, e = tupla # Se asigna los valores de la tupla a su correspondiente variable
print(a, b, c, d, e)
print('-----------------')

print('Punto 3')
# tupla_d NO se puede MODIFICAR es INMUTABLE. Pero se puede crear una lista copia y crear una nueva tupla con el valor modificado
tupla_d = (11, [22, 33], 44, 55)
tupla_d[1][0] = 222
print(tupla_d)
````
